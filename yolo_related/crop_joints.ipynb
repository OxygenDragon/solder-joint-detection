```python
from os import listdir
from PIL import Image
from os.path import isfile, join

rcnn_ann_url = './rcnn_ann.csv'
class_ann_url = './class_ann.csv'
crop_url = './crop_img/'
crop_img_count = 0


f = open(rcnn_ann_url, 'r')
class_f = open(class_ann_url, 'a+')

lines = f.readlines()
for line in lines:
    (path, x_t, y_t, x_b, y_b, label) = line.split(',')
    class_ann = []
    crop_img_count += 1
    class_ann.append('/content/drive/Shareddrives/CASLab ARC/PCB/crop_img/' + str(crop_img_count) +'.jpg')
    class_ann.append(label)
    class_str = ','.join(class_ann)
    class_f.write(class_str)

    width = float(x_b) - float(x_t)
    height = float(y_b) - float(y_t)

    img = Image.open(path)
    w, h = img.size
    im_crop = img.crop((
        float(x_t)-width*0.2 if float(x_t)-width*0.2 > 0 else 0, 
        float(y_t)-height*0.2 if float(y_t)-height*0.2 > 0 else 0, 
        float(x_b)+width*0.2 if float(x_b)+width*0.2 < w else w, 
        float(y_b)+height*0.2 if float(y_b)+height*0.2 < h else h
    ))
    im_crop.save(crop_url + str(crop_img_count) + '.jpg', "JPEG")
    print(str(crop_img_count) + '.jpg')



```
