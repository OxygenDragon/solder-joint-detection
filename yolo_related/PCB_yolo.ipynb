```python
#split data, remove all related files before started
! rm /home/user/PCB/PCB/training_set/*
! rm /home/user/PCB/PCB/valid_set/*
! rm /home/user/PCB/PCB/cfg/train.txt
! rm /home/user/PCB/PCB/cfg/valid.txt

from os import listdir
from os.path import isfile, join
import random

data_url = '/home/user/PCB/slice_img/'
annotation_url = '/home/user/PCB/slice_annotation/'

onlyfiles = [f for f in listdir(data_url) if isfile(join(data_url, f))]
random.shuffle(onlyfiles)
split_point = int(len(onlyfiles) * 0.8)
train_set = onlyfiles[:split_point]
valid_set = onlyfiles[split_point:]

f = open("/home/user/PCB/PCB/cfg/train.txt", "a+")
for file_name in train_set:
    if file_name[:-4] + ".txt" in listdir(annotation_url):
        f.write("/home/user/PCB/PCB/training_set/" + file_name + "\n")
f.close()

f = open("/home/user/PCB/PCB/cfg/valid.txt", "a+")
for file_name in valid_set:
    if file_name[:-4] + ".txt" in listdir(annotation_url):
        f.write("/home/user/PCB/PCB/valid_set/" + file_name + "\n")
f.close()
```

```python
# move to training position
import os
import shutil

training_url = "/home/user/PCB/PCB/training_set/"
valid_url = "/home/user/PCB/PCB/valid_set/"

f = open("/home/user/PCB/PCB/cfg/train.txt", 'r')
lines = f.readlines()
for line in lines:
    print("Now copying:", line)
    split = line.split('/')
    shutil.copyfile((data_url+split[-1]).strip(), training_url+split[-1].strip())
    shutil.copyfile((annotation_url+split[-1][:-5]+".txt"), training_url+split[-1][:-5]+".txt")
f.close()

f = open("/home/user/PCB/PCB/cfg/valid.txt", 'r')
lines = f.readlines()
for line in lines:
    print("Now copying:", line)
    split = line.split('/')
    shutil.copyfile((data_url+split[-1]).strip(), valid_url+split[-1].strip())
    shutil.copyfile((annotation_url+split[-1][:-5]+".txt"), valid_url+split[-1][:-5]+".txt")
f.close()
    
```

```python
# change to grey scale
from PIL import Image

img_files = [f for f in listdir(training_url) if f[-4:]==".jpg" or f[-4:]==".png"]

for img_url in img_files:
    print(training_url+img_url)
    img = Image.open(training_url+img_url).convert('L')
    img.save(training_url+img_url)

img_files = [f for f in listdir(valid_url) if f[-4:]==".jpg" or f[-4:]==".png"]

for img_url in img_files:
    print(valid_url+img_url)
    img = Image.open(valid_url+img_url).convert('L')
    img.save(valid_url+img_url)
```
